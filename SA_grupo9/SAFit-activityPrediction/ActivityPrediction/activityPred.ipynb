### Dados Numéricos

#### HRmean

```python
''' Detection '''
# IQR
Q1 = np.percentile(df['HRmean'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['HRmean'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df.shape)

''' Removing the Outliers '''
df = df[(df['HRmean'] <= (Q3+1.5*IQR)) & (df['HRmean'] >= (Q1-1.5*IQR))]
 
print("New Shape: ", df.shape)
```

## Modelos

#### Separação Treino Teste

```python
df.info()
```

```python
df = df.drop(['Activity'], axis=1)

# REMOÇÃO DOS MISSING VALUE
X = df.drop(['ActivityEncoded'], axis=1)
y = df['ActivityEncoded'].to_frame()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=2022)
```

```python
sns.distplot(y_train)
```

### Regressão Logística

#### Treino e Previsão

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
logmodel = make_pipeline(StandardScaler(), LogisticRegression(solver='saga', max_iter=5000, random_state=2021))
params= {'logisticregression__C':[0.001,.009,0.01,.09,1,5,10,25], 'logisticregression__penalty':['l1', 'l2']}
```

```python
grid = GridSearchCV(logmodel, params, cv=KFold(n_splits=10), n_jobs=1, verbose=2, return_train_score=True, scoring='accuracy', refit=True)
```

```python
grid.fit(X_train,y_train.values.ravel())
```

```python
predictions = grid.predict(X_test)
predictions
```

#### Avaliação

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
print(classification_report(y_test, predictions))
```

```python
ax = plt.subplot()

plot_confusion_matrix(grid, X_test, y_test, ax=ax)
#Deepsleep     500
#Lightsleep    500
#Other         500
#Walking       500

ax.xaxis.set_ticklabels(['Deepsleep', 'Lightsleep', 'Other', 'Walking'])
ax.yaxis.set_ticklabels(['Deepsleep', 'Lightsleep', 'Other', 'Walking'])

plt.show()
```

```python
print('Accuracy: %.3f' %accuracy_score(y_test,predictions))
```

### Redes Neuronais Artificiais

#### Treino e Previsão

```python
import tensorflow as tf
import tensorflow.keras.layers as L
```

```python
def build_model(num_classes, activation='relu', loss='sparse_categorical_crossentropy'):
    model = tf.keras.Sequential()
   
    model.add(L.Dense(32, input_dim=num_classes, activation='relu'))
    model.add(L.Dense(64, activation='relu'))
    model.add(L.Dense(64, activation='relu'))
    model.add(L.Dense(128, activation='relu'))
    model.add(L.Dropout(0.25))
    model.add(L.Dense(4, activation='softmax'))
   
    
    model.compile(loss=loss, optimizer = 'adam', metrics = ['accuracy'])
    
    return model
```

```python
model = build_model(4, activation='relu', loss='sparse_categorical_crossentropy')
model.summary()
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test,y_test),
    batch_size=64,
    epochs=100,
    verbose=1
)
```

```python
model.evaluate(X_test, y_test)
```
